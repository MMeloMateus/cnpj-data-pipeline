```python
import zipfile
import os

# Resolve o caminho absoluto do ZIP
zip_filename = os.path.abspath("Cnaes.zip")

path = r"C:\Users\mateu\Documents\GitHub\cnpj-data-pipeline\data\bronze\2023-12"
file_to_extract = "F.K03200$Z.D31209.CNAECSV"

# Garante que o diret√≥rio de destino existe
os.makedirs(path, exist_ok=True)

# Valida√ß√£o extra (boa pr√°tica)
if not os.path.isfile(zip_filename):
    raise FileNotFoundError(f"ZIP n√£o encontrado em: {zip_filename}")

with zipfile.ZipFile(zip_filename, "r") as archive:
    if file_to_extract in archive.namelist():
        archive.extract(member=file_to_extract, path=path)
        print(f"Arquivo '{file_to_extract}' extra√≠do com sucesso.")
    else:
        print(f"Arquivo '{file_to_extract}' n√£o encontrado no ZIP.")

print(f"Processo finalizado para {zip_filename}")
```

```python
import zipfile
import os

zip_filename = r"C:\Users\mateu\Documents\GitHub\cnpj-data-pipeline\data\bronze\Cnaes.zip"
path = r"C:\Users\mateu\Documents\GitHub\cnpj-data-pipeline\data\bronze\2023-12"

os.makedirs(path, exist_ok=True)

# 1Ô∏è‚É£ Tenta ler o ZIP (se existir)
try:
    with zipfile.ZipFile(zip_filename, "r") as archive:
        archive.extractall(path)
        print("ZIP lido e arquivos extra√≠dos.")
except (FileNotFoundError, zipfile.BadZipFile) as e:
    print("‚ö†Ô∏è N√£o foi poss√≠vel ler o ZIP.")
    print(f"Motivo: {e}")
    print("‚û°Ô∏è Continuando com os arquivos j√° existentes na pasta.")

# 2Ô∏è‚É£ L√™ TODOS os arquivos presentes na pasta
files_in_folder = [
    os.path.join(path, f)
    for f in os.listdir(path)
    if os.path.isfile(os.path.join(path, f))
]

if files_in_folder:
    print("‚úÖ Arquivos dispon√≠veis para processamento:")
    for f in files_in_folder:
        print("-", f)
else:
    print("‚ùå Nenhum arquivo encontrado na pasta.")
```

```python
import zipfile
import os

path = r"C:\Users\mateu\Documents\GitHub\cnpj-data-pipeline\data\bronze\2023-12"
extract_path = path  # pode ser outra pasta, se quiser

os.makedirs(extract_path, exist_ok=True)

zip_files = [
    os.path.join(path, f)
    for f in os.listdir(path)
    if f.lower().endswith(".zip")
]

print(f"üì¶ ZIPs encontrados: {len(zip_files)}")

for zip_file in zip_files:
    try:
        with zipfile.ZipFile(zip_file, "r") as archive:
            archive.extractall(extract_path)
            print(f"‚úÖ Extra√≠do: {os.path.basename(zip_file)}")
    except zipfile.BadZipFile:
        print(f"‚ùå ZIP corrompido ou inv√°lido: {zip_file}")

print("üèÅ Processamento finalizado.")
```
